# Summary

The live and geocarbon above ground biomass datasets are sparse for many areas, except for the **Central and Southern Cascades forests**, **Central Pacific coastal forests** and **Queen Charlotte Islands**. These available data may be useful as a reference to be compared with results of simulation runs.

# Analysis

```python
holdridge = Shape('HoldridgeLifeZones.json', 'l')
```

```python
emerald = Shape('emerald.shp', 'l') 
```

```python
emerald.get_bounds()
```

```python
# N40-80, W140-180
n80w180 = Layer(r'N80W180_agb.tif')
```

```python
# N40-80, W120-140
n80w140 = Layer(r'N80W140_agb.tif')
```

```python
geocarbon = Layer(r'GEOCARBON_AGB_Map180.tif')
```

### Biomass values from available data (Land Data Set)

```python
clip_raster(n80w180.img, emerald.df, 'emerald_n80w180')
clip_raster(n80w140.img, emerald.df, 'emerald_n80w140')
```

```python
emerald_n80w180 = Layer(r'emerald_n80w180.tif')
emerald_n80w140 = Layer(r'emerald_n80w140.tif')
```

```python
# N40-80, W120-140
plot_raster(emerald.df, emerald_n80w140.img, 0.000001, 'ABOVEGROUND LIVE BIOMASS', 'Spectral_r')
```

```python
clip_raster(geocarbon.img, emerald.df, 'emerald_geocarbon')
```

```python
emerald_geocarbon = Layer(r'emerald_geocarbon.tif')
```

```python
plot_raster(emerald.df, emerald_geocarbon.img, 0.000001, 'ABOVEGROUND GEOCARBON BIOMASS', 'Spectral_r')
```

More data are available for Central and Southern Cascades forests, Central Pacific coastal forests and Queen Charlotte Islands.

### Central and Southern Cascades forests

```python
censud = Shape('Central and Southern Cascades forests.shp', 'l')
```

```python
censud_holdridge_df = holdridge.df.clip(censud.df)
```

```python
censud_inventory = inventorize(censud_holdridge_df)
```

```python
plot_vector(censud_inventory, 'LifeZone', 'LIFEZONES', 'Spectral')
```

```python
clip_raster(n80w140.img, censud.df, 'cenpac_aglb')
clip_raster(geocarbon.img, censud.df, 'cenpac_aggb')
```

```python
censud_aglb = Layer('censud_aglb.tif')
censud_aggb = Layer('censud_aggb.tif')
```

```python
plot_raster(censud.df, censud_aglb.img, 0.000001, 'ABOVEGROUND LIVE BIOMASS', 'Spectral_r')
```

```python
censud_aglb_df = raster_dataframe(censud_aglb.img, censud_inventory, 'Live')
```

```python
plot_raster(censud.df, censud_aggb.img, 0.000001, 'ABOVEGROUND GEOCARBON BIOMASS', 'Spectral_r')
```

```python
censud_aggb_df = raster_dataframe(censud_aggb.img, censud_inventory, 'Geo')
```

```python
censud_agb_df = pd.merge(censud_aglb_df, censud_aggb_df, on='LifeZone')

censud_agb_df
```
